### cv2.imread()

```python
im = cv2.imread('pic_resources/pic_resources/lenna.png')
```

### plt.imshow()

```python
plt.imshow(im)
```

### cv2.cvtColor
BGR -> RGB

```python
rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

BGR -> gray

```python
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

crop 切り抜き

```python
crop = im[50:-50, 50:-50, :]
plt.imshow(crop)
```

### cv2.imwrite()

```python
cv2.imwrite('pic_resources/pic_resources/cropped_lenna.png', crop)
```

## 65.OpenCVで画像処理をしてみる

### Binarization (2値化)
- cv2.threshold()
- cv2.adaptiveThreshold()

cv2.threshold()

```python
im = cv2.imread("pic_resources/pic_resources/lenna.png")
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
retval, dist = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(dist, cmap='gray')
```

cv2.adaptiveThreshold()

```python
im = cv2.imread("pic_resources/pic_resources/text_pic.jpg")
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
dst = cv2.adaptiveThreshold(
    gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 4)
```

```python
plt.imshow(dst, cmap='gray')
```
