色分け

```python
# vote_averageが下位50%だったらlow、上位だったらhigh
# df['evaluation'] = ~~

df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df = df[df['vote_count'] > 100]

df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x >= df['vote_average'].median() else 'low')
```

```python
sns.pairplot(df[['revenue', 'budget', 'vote_average', 'vote_count', 'evaluation']],
hue = 'evaluation',
plot_kws={'alpha':0.3},
diag_kind='hist')
```
