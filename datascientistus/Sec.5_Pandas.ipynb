## 42.表結合
- pd.concat([df1,df2,...], axis=0 or 1) axis=1(列を結合)は余り使わない
- df1.merge(df2)
  - 共通のカラムをキーに結合
  - howという引数でどのように結合するかを指定
    - how=left, right, inner, outer
  - onという引数で結合のキーとなるカラムを指定
  - left_on, right_on　という引数でキー名が異なる場合でも結合できる
  - suffixes = _str で結合後のカラムの名前を変えられる

```python
df1 = pd.DataFrame(
    {"Key": ["k0", "k1", "k2"],
     "A": ["a0", "a1", "a2"],
     "B": ["b0", "b1", "b2"]}
)
df2 = pd.DataFrame(
    {"Key": ["k0", "k1", "k2"],
     "C": ["c0", "c1", "c2"],
     "D": ["d0", "d1", "d2"]}
)
```

```python
df1
```

```python
df2
```

### .merge()
共通のカラムをkeyに結合

```python
df1.merge(df2)
```

```python
df1 = pd.DataFrame(
    {"Key": ["k0", "k1", "k2"],
     "A": ["a0", "a1", "a2"],
     "B": ["b0", "b1", "b2"]}
)
df2 = pd.DataFrame(
    {"Key": ["k0", "k1", "k3"],
     "C": ["c0", "c1", "c3"],
     "D": ["d0", "d1", "d3"]}
)
```

```python
df1.merge(df2, how='inner', on='Key')
```

```python
df1.merge(df2, how='outer', on='Key')
```

```python
df1.merge(df2, how='left', on='Key')
```

```python
df1 = pd.DataFrame(
    {"Key1": ["k0", "k1", "k2"],
    "ID": ["aa", "bb", "cc"],
     "A": ["a0", "a1", "a2"],
     "B": ["b0", "b1", "b2"]}
)
df2 = pd.DataFrame(
    {"Key2": ["k0", "k1", "k3"],
    "ID": ["aa", "bb", "cc"],
     "C": ["c0", "c1", "c3"],
     "D": ["d0", "d1", "d3"]}
)
```

### .apply()
- 関数を各行に適用する
- あるカラムと別のカラムの計算などに使う

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
# 収益 = revenue - budget
df.head(1)
```

```python
# budgetが0の行をnanにする
# return_nan_if_zero(0) --> nan
# return_nan_if_zero(100000) --> 100000

def return_nan_if_zero(budget):
    if budget == 0:
        return np.nan
    else:
        return budget
    
return_nan_if_zero(0)
```

```python
df['budget'].apply(return_nan_if_zero)
```

```python
df['budget'] = df['budget'].apply(return_nan_if_zero)
```

```python
df
```

```python
# def return_nan_if_zero(budget):
#     if budget == 0:
#         return np.nan
#     else:
#         return budget

z = lambda budget: np.nan if budget == 0 else budget
z(0)
```

```python
df['budget'] = df['budget'].apply(lambda budget: np.nan if budget == 0 else budget)
df['revenue'] = df['revenue'].apply(lambda budget: np.nan if budget == 0 else budget)
```

```python
df
```

```python
df = df.dropna(subset=['budget', 'revenue'])
```

```python
df
```

```python
# profit = revenue - budget
df['profit'] = df.apply(lambda row:row['revenue'] - row['budget'], axis=1)
```

```python
df
```

```python
df.sort_values('profit', ascending=False).head(1)
```

```python
# profitが最大となる行を抽出
df.iloc[df['profit'].idxmax()]
```

```python
# profitが最大となる行を抽出する方法2
df[df['profit'] == df['profit'].max()]
```

## 46.DataFrameの保存

### .to_csv()

```python
### profit (revenue - budget) の高い順に並べた表を作成してみる ( budget やrevenueが0のものは除く)

# datasetの読み込み
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df
```

```python
## budget と revenue が0のものをnanにする
#df['budget'] = df['budget'].apply(lambda x: np.nan if x == 0 else x)
#df['revenue'] = df['revenue'].apply(lambda x: np.nan if x == 0 else x)

## budget と revenue がNanの行を除去する
#df = df.dropna(subset=['budget', 'revenue'])

## より簡単な方法
df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]

## profit を計算する
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)

## profit を高い順にソートする
df = df.sort_values('profit', ascending=False)
df
```

```python
# 元のindexは不要なので、オプションで除去
df.to_csv('dataset/tmdb_5000_movies_profit_sorted.csv', index=False)
```
