# OpenCV

```python
import cv2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## 64.OpenCVを使ってみる
- cv.imread()
- plt.imshow()
- cv2.cvtColor()
  - cv2.COLOR_BGR2RGB
  - cv2.COLOR_BGR2GRAY
- crop = 画像のスライシング
- cv.imwrite()

### cv2.imread()

```python
im = cv2.imread('pic_resources/pic_resources/lenna.png')
```

### plt.imshow()

```python
plt.imshow(im)
```

### cv2.cvtColor
BGR -> RGB

```python
rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

BGR -> gray

```python
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

crop 切り抜き

```python
crop = im[50:-50, 50:-50, :]
plt.imshow(crop)
```

### cv2.imwrite()

```python
cv2.imwrite('pic_resources/pic_resources/cropped_lenna.png',crop)
```

```python
crop_load = cv2.imread('pic_resources/pic_resources/cropped_lenna.png')
```

```python
plt.imshow(crop_load)
```
